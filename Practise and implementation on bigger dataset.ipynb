## Importing stop words from nltk

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stop_words=set(stopwords.words('english'))
```

```python
stop_words
```

## Sentence Tokenization

```python
from nltk.tokenize import sent_tokenize
```

```python
nltk.download('punkt')
```

```python
tokenized_text=sent_tokenize(data)
```

```python
tokenized_text
```

### Word Tokenization

```python
from nltk.tokenize import word_tokenize
tokenized_word=word_tokenize(data)
print(tokenized_word)
```

```python
from nltk.probability import FreqDist
fdist = FreqDist(tokenized_word)
```

```python
from matplotlib.pyplot import figure
figure(figsize=(8, 6), dpi=80)

fdist.plot(30,cumulative=False)
plt.show()
```

## Removing Stop words 

```python
filtered_sent=[]
for w in tokenized_word:
    if w not in stop_words:
        filtered_sent.append(w)
print("Filterd Sentence:",filtered_sent)
```

```python
fdist2 = FreqDist(filtered_sent)
```

```python
figure(figsize=(8, 6), dpi=80)

fdist2.plot(30,cumulative=False)
plt.show()
```

### As we can see that punctuations are there let's remove them

```python
import string
def remove_punctuation(t):
    no_punct = ' '.join(x for x in t if x not in string.punctuation)
    return no_punct
```

```python
filtered_sent2 = remove_punctuation(filtered_sent)
```

```python
removing_missing_punc = ["!", "--", " ` "]
```

```python
filtered_sent3 = ' '.join(x for x in filtered_sent2 if x not in removing_missing_punc)
```

```python
filtered_sent2.split()
```

```python
fdist4 = FreqDist(filtered_sent2.split()[4:])
figure(figsize=(8, 6), dpi=80)
fdist4.plot(30,cumulative=False)
plt.show()
```

## Performing Lemmatization

```python
nltk.download('wordnet')
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
lem = WordNetLemmatizer()

Lemmatization_words=[]

for w in filtered_sent2.split():
    Lemmatization_words.append(lem.lemmatize(w))
```

```python
Lemmatization_words
```

```python
fdist3 = FreqDist(Lemmatization_words)
```

```python
figure(figsize=(8, 6), dpi=80)

fdist3.plot(30,cumulative=False)
plt.show()
```
